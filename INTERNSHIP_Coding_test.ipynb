#### Write a function that takes OfferDetails as input, and returns FaceValue as output

```python
def get_FaceValue(string_S):   
    y = 'NO _Valu'
    if re.findall('¢',string_S) or re.findall('\$',string_S): # Finding the element which contain dollar or cent symbol 
        for x in string_S.split():
            if ('$' in x) or ('¢'in x): 
                return x
    else:
        return np.nan # Return the Fave Value
        
```

```python
# Check the function functionality 
print(get_FaceValue(df[4:5].to_string()))
print(get_FaceValue(df[880:881].to_string()))
```

#### Run the function against all rows in the attached data and determine function accuracy (Higher the better)

```python
for x in range(1,df.size+1):
    #get_FaceValue(df[:x].to_string())
    print(x,get_FaceValue(df[x-1:x].to_string()))
```

```python
df1=df.copy()
```

```python
df1['FaceValue']=df['Offer details'].apply(lambda x: get_FaceValue(x))
```

We can notice from the dataset that, we have to buy particular amount/quantity of products to avail the offer i.e. our Purchase Requirement.

```python
def quantities(doc):
    doc=nlp(doc)
    list2 = []
    for ent in doc.ents:
          if ent.label_ in {'CARDINAL'} :
            list2.append(ent.text)
    return list2
```

```python
quantities(str3)
```

```python
for x in range(1,df.size+1):
    print(x,quantities(df[x-1:x].to_string()))
```

### Excercise 3
#### Write a function that takes OfferDetails as input, and returns Product as output (Product can be single or array)

```python
str3='Save $2.00 ONE Tide Laundry Detergent (excludes Tide Purclean Laundry Detergent, Studio by Tide Laundry Detergent, Tide PODS, Tide Rescue, Tide Simply Laundry Detergent, Tide Simply PODS, Tide Detergent 10 oz and trial/travel size).	'
```

```python
doc2=nlp(str3)
displacy.render(doc2, style="ent")
```

```python
def product_names(doc):
    doc=nlp(doc)
    list1 = []
    for ent in doc.ents:
          if ent.label_ in {'ORG','LOC','WORK_OF_ART','EVENT','PRODUCT',"NORP",'PERSON','GPE','FAC'} :
            list1.append(ent.text)
    return list1
```

```python
product_names(str3)
```

### Write a function that takes OfferDetails as input, and returns Product as output (Product can be single or array)

```python
for x in range(1,df.size+1):
    print(x,product_names(df[x-1:x].to_string()))
```

```python
df1['Product']=df1['Offer details'].apply(lambda x: product_names(x))
```

Accuracy of this function is around 75%.
